```python
import csv
import sys, os, re
import json
import codecs

with open('QA_Immigration.csv', newline='') as csvfile:
    qareader = csv.DictReader(csvfile)
    path = "./test.jsonl"
    f = codecs.open(path, 'w', 'utf-8')
    for row_object in qareader:
        # ensure_ascii=False is essential or errors/corruption will occur
        if row_object['问题'] != '':
            # SKip the line without questions but still keep the line with question but no answers
            json_record = json.dumps(row_object, ensure_ascii=False)
            f.write(json_record + "\n")

    f.close()
```
